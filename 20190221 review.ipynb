```python
import pandas as pd
df = pd.DataFrame()
print(df, "\n")
df2 = pd.DataFrame([1,2,3,4])
print(df2, "\n")
df3 = pd.DataFrame([[1,2,3,4],[5,6,7,8]])
print(df3, "\n")
df4 = pd.DataFrame([[1,2,3,4],[5,6,7,8]], index = ['one', 'two'], columns=['a','b','c','d'])
print(df4, "\n")


# Column Selection
print(df4['a'], "\n")
print(df4[['a', 'b']], "\n")
print(type(df4['a']), "\n")
df4['e'] = [1,2] # add data
print(df4, "\n")


del df4['d'] # delete data
print(df4, "\n")
print(df4.iloc[0,0], "\n")
print(df4.iloc[:,0], "\n")
print(df4.iloc[:,[0,2,3]], "\n")
print(df4.loc[:,['b','c','e']], "\n")
```
