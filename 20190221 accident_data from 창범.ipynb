```python
import pandas as pd

df = pd.read_csv("carAccident.csv",encoding='EUC-KR')
need = ['관할경찰서','발생건수','사상자수','사망자수','중상자수','경상자수','부상신고자수']

df_need= df[need]

df_need_grouped = df_need.groupby('관할경찰서').sum()

df_need_se= df_need[df_need['관할경찰서'].str.contains('서울특별시')].sum()
df_need_bu= df_need[df_need['관할경찰서'].str.contains('부산광역시')].sum()
df_need_dg= df_need[df_need['관할경찰서'].str.contains('대구광역시')].sum()
df_need_ic= df_need[df_need['관할경찰서'].str.contains('인천광역시')].sum()
df_need_us= df_need[df_need['관할경찰서'].str.contains('울산광역시')].sum()
df_need_gj= df_need[df_need['관할경찰서'].str.contains('광주광역시')].sum()
df_need_gg= df_need[df_need['관할경찰서'].str.contains('경기도')].sum()
df_need_gy= df_need[df_need['관할경찰서'].str.contains('강원도')].sum()
df_need_cb= df_need[df_need['관할경찰서'].str.contains('충청북도')].sum()
df_need_cn= df_need[df_need['관할경찰서'].str.contains('충청남도')].sum()
df_need_jb= df_need[df_need['관할경찰서'].str.contains('전라북도')].sum()
df_need_jn= df_need[df_need['관할경찰서'].str.contains('전라남도')].sum()
df_need_gb= df_need[df_need['관할경찰서'].str.contains('경상북도')].sum()
df_need_gn= df_need[df_need['관할경찰서'].str.contains('경상남도')].sum()

local=['서울특별시','부산광역시','대구광역시','인천광역시','울산광역시','광주광역시','경기도','강원도'
      ,'충청북도','충청남도','전라북도','전라남도','경상북도','경상남도']

df_final = pd.concat([df_need_se,df_need_bu,df_need_dg,df_need_ic,df_need_us,df_need_gj,df_need_gg,df_need_gy,
                      df_need_cb,df_need_cn,df_need_jb,df_need_jn,df_need_gb,df_need_gn],axis=1,keys=local)

df_final=df_final[1:]
df_final

df_final.plot(kind='line')
df_final.plot(kind='bar')
```
