```python
import os
dirpath = "Text-books"
files = os.listdir(dirpath)

context = ""

for file in files:
    c = open((dirpath + "\\" + file), "r", encoding='utf-8-sig')
    context = context + c.read()
    
context = context.split()

print(context)
```

```python
dic = {}

for i in context:
    if i not in dic:
        dic[i] = 0
    else :
        dic[i] = dic[i] +1

        
        
#print(dic)
import operator
sortedArr = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
arr = sortedArr[0:10]
arr


import pandas as pd

p = pd.DataFrame(arr)
p.columns = ['a','b']
import matplotlib.pyplot as plt
p.index =  p['a']
del p['a']
plt.plot(p)
```
