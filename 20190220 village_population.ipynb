```python
#송도 1동 이라는 이름이 포함된 지역의 인구 그래프를 그리시오

import csv

f = open("village_population.csv", encoding="UTF-8")
data = csv.reader(f)
next(data)

songdo = []

for row in data:
    if "송도1동" in row[0]:
        for r in row[3:]:
            songdo.append(int(r))

print(songdo)

import matplotlib.pyplot as plt

plt.plot(songdo)
plt.title("songdo 1dong pop")
plt.xlabel("age")
plt.ylabel("pop")
plt.show()

                          
plt.hist(songdo)
plt.title("songdo 1dong pop")
plt.xlabel("age")
plt.ylabel("pop")
plt.show()
```
