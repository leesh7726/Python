```python
#<<<<< 1부터 100까지의 합을 구현하시오. >>>>>
sum=0
for i in range(1,101):
    sum+=i
    
print(sum)

#<<<<< 1부터 100까지의 홀수의 합을 구현하시오. >>>>>
sum=0
for i in range(1,101,2):
    sum+=i
    
print(sum)


#<<<<< 1부터 100까지의 짝수의 합을 구현하시오. >>>>>
sum=0
for i in range(2,101,2):
    sum+=i
    
print(sum)

#<<<<< 시작값, 끝값, 증가값 3가지 입력을 받아서 합계를 처리하는 기능을 구현하시오. >>>>>
start = int(input("시작값 입력"))
end = int(input("끝값 입력"))
plus = int(input("증가값 입력"))

sum = 0
for i in range(start, end+1, plus):
    sum+=i
    

print(sum)
```
