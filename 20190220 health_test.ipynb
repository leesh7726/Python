```python
import numpy as np
data = np.loadtxt(fname = "health_test_info.csv",  delimiter=',', dtype=np.str)
#data.shape
```

```python
height = data[0:,5]
#0 > 180  // 1> 175 // 2> 170 // 3 > 165 // 4> 160 // 5> 150 
cnt = np.zeros(7, dtype=np.int)
height == '180'


for h in height[1:] :
    if h == '180':
        cnt[0] += 1
    if h == '175':
        cnt[1] += 1
    if h == '170':
        cnt[2] += 1
    if h == '165':
        cnt[3] += 1
    if h == '160':
        cnt[4] += 1
    if h == '155':
        cnt[5] += 1
    if h == '150':
        cnt[6] += 1
        
print(cnt)
title = ['180', '175', '170', '165', '160', '155', '150']
import matplotlib.pyplot as plt
plt.plot(cnt)
plt.show()

plt.bar(title , cnt)
plt.show()

```
