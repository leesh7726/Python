```python
import numpy as np
import pandas as pd

df = pd.read_csv("seoul.csv", encoding='utf-8')
data = np.array(df)


for row in data:
    if float(row[3]) == float(df['최저기온(℃)'].min()) :
        print("날짜 : " , row[0], "최저기온 : ", row[3])
```
