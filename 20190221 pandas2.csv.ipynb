```python
import pandas as pd
df = pd.read_csv("seoul.csv", encoding='utf-8')
print(df.head()) #처음 5개 출력
print()

print(df.tail()) #마지막 5개 출력
print()

print(df['날짜'].head()) #칼럼으로 검색
print()

print(df['지점'].head())
print()

print(df[['지점','날짜']].head())
print()

print(df.columns)  # 칼럼만 출력
print(df.columns[0:2]) #칼럼 slicing
print()

print(df[1:4], "\n")  # slicing, 하지만 index로 하나씩 못가져옴
print(df.iloc[0:2], "\n")# slicing. index로 하나씩 가져올 수도 있음 @@@@@indexlocation???@@@@@
print(df.iloc[2], "\n") # index
print(df.iloc[2:7, 0:3], "\n") # 특정 범위 출력
print(df.iloc[[0,2,4], 0:3], "\n") # 특정 행만 가져오기

print()


print(df.iloc[0,0],"\n")
print(df.loc[2,'날짜'], "\n")

print(df.iloc[[0,2,4], 0:2],"\n") # iloc -> index로만 가능
print(df.loc[[0,2,4],['날짜','지점']]) # loc -> 문자로만 가능
```
