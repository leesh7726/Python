```python
# plot
import numpy as np
x = np.arange(0,12,0.01)
x.shape
y = np.sin(x)
y1 = np.cos(x)

plt.figure(figsize=(10,4)) # Size 조절
plt.grid() # 눈금


plt.xlabel("x axis") # label
plt.ylabel("y axis") # label
plt.title("x-y title") # title

plt.ylim(-1.5,1.5) # 범위 지정 (혹은 확대?)
plt.xlim(0, np.pi) # 범위 지정(혹은 확대?)

plt.plot(x, y, label="sin")
plt.plot(x, y1, label="cos")
plt.legend() #출력 그래프 구별 목록 // plt.plot()안에 label을 먼저 써주어야 한다

plt.show()
```

```python
t = np.arange(0,5,0.5)
plt.figure(figsize=(10,6))
plt.plot(t,t,'r--') # 'r--' ->  빨간점선 // 'b--'  -> 파란점선
plt.plot(t,t+10, color = 'purple', linestyle='dashed') # 보라색 점선
plt.plot(t,t*15, color = 'pink', linestyle='dashed', marker='o') #분홍색 점선, 동그라미마커
plt.plot(t,t**2+5, color = 'black', linestyle='dashed', marker='o', markerfacecolor='white', markersize=10) 
plt.plot(t, t**2, 'bs') # 'bs' -> 파랑네모모양 점 s -> 사각형점
plt.plot(t, t**3, 'g^') # 'g^' -> 초록세모모양 점 ^ -> 세모점
plt.plot(t, t*10, 'yx') # 'bs' -> 노랑x모양 점 x -> X점

plt.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1,s2,s3))
plt.show()
```
