```python
#교통사고정보를 이용하여 각각의 시도별 발생건수, 사상자수, 사망자수, 중사자수, 경상자수, 부상신고자수 합계를 구하고
#그래프로 출력하시오(pandas or numpy 사용)
# => 관할경찰서
# => 시도별(서울특별시, ...)

import pandas as pd
import numpy as np
df = pd.read_csv("carAccident.csv", encoding='EUC-KR')
data = pd.DataFrame(df.loc[:,'관할경찰서' : '부상신고자수'])

data = np.array(data)



dic = {}
for row in data:
    city = row[0].split()[0]
    if city not in dic:
        dic[city] = np.zeros(6)
    
    else :
        dic[city] = dic[city] +  row[2:8]
        
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc #한글 깨지는거 방지
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

titles = ['발생건수','사상자수','사망자수','중상자수','경상자수','부상신고자수']
for city in dic.keys():
    #plt.plot(titles, dic[city])
    plt.bar(titles, dic[city])
    print(city)
    plt.show()
```
