```python
import pandas as pd
import numpy as np

def add(a,b):
    return a + b

df = pd.DataFrame(np.random.rand(5,3), columns=['col1','col2','col3'])

print(df, "\n")

print(df.pipe(add,2), "\n") # 모든값에 2 더함

print(df.apply(np.mean), "\n") # columns 기준

print(df.apply(np.mean, axis = 1),"\n") # index 기준

print(df.apply(lambda x:x.max() - x.min()), "\n") # ???@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@존재하지 않는 함수 정의해서 사용???
```

```python
# DataFrame Iteration
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(5,3), columns=['col1','col2','col3'])

for col in df:
    print(col) # column만 나옴
print("\n")

for i in df.values:
    print(i)
print("\n")
    
for key, values in df.iteritems(): # index 기준
    print(key, values, "\n")
    break
print("\n")


for row_index, row in df.iterrows(): # row기준
    print(row_index, row, "\n")
    break
print("\n")

for row in df.itertuples(): # row기준
    print(row, "\n")
    break
print("\n")
```

```python
# sorting

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(5,2), index=[2,1,0,3,4])
print(df, "\n")

df = df.sort_index(ascending=False)
print(df, "\n")

df = df.sort_index(axis = 1)
print(df, "\n")

df.rename(columns = {
    df.columns[0] : 'col1',
    df.columns[1] : 'col2'}, inplace=True
)
print(df, "\n")

df = df.sort_values(by='col1')
print(df, "\n")

df = df.sort_values(by='col2')
print(df, "\n")

df = df.sort_values(by=['col1', 'col2'])
print(df, "\n")

df = df.sort_values(by=['col1', 'col2'], ascending=[True,False])
print(df, "\n")
```

```python
# Missing Data -> NA or NaN
df = pd.DataFrame(np.random.randn(5,3),
                  index = ['a','c','e','f','h'],
                  columns=['one','two','three'])
print(df, "\n")

df = df.reindex(['a','b','c','d','e','f','g','h'])
print(df, "\n")
print(df['one'].isnull(), "\n")
print(df['one'].notnull(), "\n")
print(df['one'].sum(), "\n")
print(df.dropna(), "\n") # NaN값 제거 // 0으로채우는건 fillna(0)


df = pd.DataFrame(index = [1,2,3,4,5],
                 columns=['one','two'])
print(df, "\n")
print(df.fillna(0), "\n") # NaN을 0으로 채움 // Nan값 제거는 dropna()
```

```python
import pandas as pd
import numpy as np

data = {'Team' : ['Tom', 'John','John', 'Amy'],
        'Rank' : [1,2,2,3],
        'Year': [2014, 2016, 2016, 2018],
        'Point' : [10, 15, 20, 10]
       }

df= pd.DataFrame(data)
print(df, "\n")

print(df.groupby('Team')['Point'].mean(), "\n")

print(df.groupby('Team').groups, "\n")

df = df.groupby('Team')
for name,group in df:
    print(name, "\n", group, "\n")


```

```python
# Merge & Join
#pd.merge(left,right,how='inner', on=None, left_on=None,
#        right_on=None,left_index = False, right_index=False, sort=True

import pandas as pd
import numpy as np

left = pd.DataFrame({
    'id' : [1,2,3,4,5],
    'Name' : ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'Subject_id' : ['sub1', 'sub2', 'sub4', 'sub6', 'sub5']})


right = pd.DataFrame({
    'id' : [1,2,3,4,5],
    'Name' : ['Alex2', 'Amy2', 'Allen2', 'Alice2', 'Ayoung2'],
    'Subject_id' : ['sub2', 'sub4', 'sub3', 'sub6', 'sub5']})


print (left, "\n\n", right, "\n")

pd.merge(left, right, on='id')
pd.merge(left, right, on=['id', 'Subject_id'])
print(pd.merge(left,right,on='Subject_id',how='left'), "\n") # Left Outer Join
print(pd.merge(left,right,on='Subject_id',how='right'), "\n") # Right Outer Join
print(pd.merge(left,right,on='Subject_id',how='outer'), "\n") # Full Outer Join
print(pd.merge(left,right,on='Subject_id',how='inner'), "\n") # Inner Join
```
