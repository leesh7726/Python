```python
import csv
f= open("seoul.csv", encoding="utf-8")
data=csv.reader(f)
next(data)
max_temp = 0
max_temp_date=''
for row in data:
    if row[-1] != '':
        if int(row[0].split("-")[0]) < 1955 and float(row[-1]) > max_temp :
            max_temp = float(row[-1])
            max_temp_date = row[0]
            
print(max_temp, max_temp_date)
```

```python
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)
temperature = []
for row in data:
    if (row[-1] != '' 
        and (row[0].split('-')[1] == '11' and row[0].split('-')[2] == '24')):
        temperature.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.plot(temperature)
plt.show()
```

```python
#역대 서울 기온 중 37 이상이었던 날의 날짜와 최고 기온을 출력하시오.
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)

over37 = []

for f in data:
    if f[-1] != '':
        if float(f[-1]) >= 37.0 :
            over37.append([f[0],f[-1]])
            

print(over37)

maxtemp = ["", 0]
for temp in over37:
    if float(temp[1]) > float(maxtemp[1]):
        maxtemp = temp

print("max = " , maxtemp)
```

```python
#역대 서울의 최고기온을 히스토그램을 출력하시오
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)
max=[]
for row in data:
    if(row[-1]!=""):
        max.append(float(row[-1]))
        
        
import matplotlib.pyplot as plt
plt.hist(max,bins = 100)
plt.show()
```

```python
import csv
f = open("village_population.csv", encoding='utf-8')
data = csv.reader(f)

next(data)
songdo= []
sindorim = [] 
junggok4 = []

for row in data:
    if "송도1동" in row[0]:
        for i in range(3,len(row)):
            songdo.append(int(row[i]))
    if "신도림" in row[0]:
        for i in range(3,len(row)):
            sindorim.append(int(row[i]))
    if "중곡제4동" in row[0]:
        for i in range(3,len(row)):
            junggok4.append(int(row[i]))
                          
                          
                          
import matplotlib.pyplot as plt
plt.hist(songdo)
plt.show()

plt.hist(sindorim)
plt.show()

plt.plot(songdo)
plt.plot(sindorim)
plt.plot(junggok4)
plt.show()
```
